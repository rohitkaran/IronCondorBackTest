```python
expiry = get_expiry_date(year=2021, month=2)
len(expiry)
list(expiry)

nifty_fut = get_history(symbol="NIFTY",
                        start=date(2021,2,1),
                        end=date(2021,2,28),
                        index=True,
                        futures=True,
                        expiry_date=max(list(expiry)))
# nifty_fut['Last']
```

```python
n=0
result=pd.DataFrame(columns=['date','strike','CE_sell','PE_sell','PE_Buy','CE_Buy','CE_sell_exit','PE_sell_exit','PE_buy_exit','CE_buy_exit'])
for dt in nifty_fut.index[:-1]:
    
    strike=round(nifty_fut['Last'].loc[dt]/50)*50
#     print(dt, strike)
    
    if(dt>=list(expiry)[n]):
        n=n+1
    expiry_dt=list(expiry)[n]
    nifty_opt_1_CE = get_history(symbol="NIFTY",
                        start=dt,
                        end=dt+timedelta(days=4),
                        index=True,
                        option_type='CE',
                        strike_price=strike,
                        expiry_date=expiry_dt)
#     print(nifty_opt_1_CE['Last'].iloc[0])
    nifty_opt_1_PE = get_history(symbol="NIFTY",
                        start=dt,
                        end=dt+timedelta(days=4),
                        index=True,
                        option_type='PE',
                        strike_price=strike,
                        expiry_date=expiry_dt)
#     print(nifty_opt_1_PE['Last'].iloc[0])
    
    
    nifty_opt_2_PE = get_history(symbol="NIFTY",
                        start=dt,
                        end=dt+timedelta(days=4),
                        index=True,
                        option_type='PE',
                        strike_price=(strike-300),
                        expiry_date=expiry_dt)
#     print(nifty_opt_2_PE['Last'].iloc[0])
    
    nifty_opt_2_CE = get_history(symbol="NIFTY",
                        start=dt,
                        end=dt+timedelta(days=4),
                        index=True,
                        option_type='CE',
                        strike_price=strike+300,
                        expiry_date=expiry_dt)
#     print(nifty_opt_2_CE['Last'].iloc[0])
    
    
    premium_recieved=nifty_opt_1_CE['Last'].iloc[0]-nifty_opt_2_CE['Last'].iloc[0]+nifty_opt_1_PE['Last'].iloc[0]-nifty_opt_2_PE['Last'].iloc[0]
    
    premium2=nifty_opt_1_CE['Last'].iloc[1]-nifty_opt_2_CE['Last'].iloc[1]+nifty_opt_1_PE['Last'].iloc[1]-nifty_opt_2_PE['Last'].iloc[1]
    list_res=[dt, strike,nifty_opt_1_CE['Last'].iloc[0],nifty_opt_1_PE['Last'].iloc[0],nifty_opt_2_PE['Last'].iloc[0],nifty_opt_2_CE['Last'].iloc[0],
             nifty_opt_1_CE['Last'].iloc[1],nifty_opt_1_PE['Last'].iloc[1],nifty_opt_2_PE['Last'].iloc[1],nifty_opt_2_CE['Last'].iloc[1]]
    time.sleep(1)
    print(list_res)
    
    result.loc[len(result)]=list_res
    
#     result=result.append(pd.DataFrame(list_res,columns=['date','strike','CE_sell','PE_sell','PE_Buy','CE_Buy','CE_sell_exit','PE_sell_exit','PE_buy_exit','CE_buy_exit']))
```

```python
result['premium_received']=result['CE_sell']+result['PE_sell']-result['PE_Buy']-result['CE_Buy']
result['Squared']=result['CE_sell_exit']+result['PE_sell_exit']-result['PE_buy_exit']-result['CE_buy_exit']
result['PNL']=result['premium_received']-result['Squared']
result
```
